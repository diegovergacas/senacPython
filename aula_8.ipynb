```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
```

Requests

```python
# Biblioteca Requests

# A biblioteca á formada por quatro principais métodos
# PUT -> atulizar
# DELETE -> excluir 
```

```python
# Busda de CEP
cep = requests.get("https://cep.awesomeapi.com.br/json/04345020")
print(cep.text)
```

```python
#200 encontrado
#400 falha na requisição
#404 não encontrado

if cep.status_code == 200:
  print(cep.text)
elif cep.status_code == 400:
  print("CEP não localizado")
else:
  print("Ocorreu um erro na requisição")
```

```python
# Fazer o Python ler arquivo Json

import json

#convertendo um arquivo json em dicionario pyhton
logradouro = json.loads(cep.text)

#imprimindo um dicionario em python
print("Dicionario python: ", logradouro)
print("Formato Json: ", cep.text)
```

```python
#Recuperando informações de um dicionario

print(logradouro["cep"])
print(logradouro["address_type"] + ' ' + logradouro["address_name"])
```

Cotação

```python
# Busca de cotação
cotacaoUSD = requests.get("https://economia.awesomeapi.com.br/last/USD-BRL")
print(cotacaoUSD.text)
cotacaoEUR = requests.get("https://economia.awesomeapi.com.br/last/EUR-BRL")
print(cotacaoEUR.text)
cotacaoBTC = requests.get("https://economia.awesomeapi.com.br/last/BTC-BRL")
print(cotacaoBTC.text)
```

```python
#from IPython.utils.path import target_update
import json
#convertendo um arquivo json em dicionario pyhton
USD = json.loads(cotacaoUSD.text)
EUR = json.loads(cotacaoEUR.text)
BTC = json.loads(cotacaoBTC.text)

#imprimindo um dicionario em python
#print(USD)

#Recuperando informações de um dicionario

#valorUSD = (USD["USDBRL"]["code"]) + " " + (USD["USDBRL"]["high"])

#print(USD["USDBRL"]["code"])
#print(USD["USDBRL"]["high"])


valorUSD = float(USD["USDBRL"]["high"])
valorEUR = eval(EUR["EURBRL"]["high"])
valorBTC = eval(BTC["BTCBRL"]["high"])
#print(valorUSD)
type(valorUSD)
#print("USD 1,00 é igual a R$ " + (USD["USDBRL"]["high"]))

print(f'USD$ 1,00 é igual a R$ {valorUSD:,.2f}')
print(f'EUR$ 1,00 é igual a R$ {valorEUR:,.2f}')
print(f'BTC$ 1,00 é igual a R$ {valorBTC:,.2f}')
```

```python
#busca a catação do dólar

cotacaoDolar = requests.get('https://www.google.com/finance/quote/USD-BRL?hl=pt')

soup = BeautifulSoup(cotacaoDolar.text, 'html.parser')

valorDolar = soup.find(class_ = "kf1m0").text

dataHora = datetime.now()

print("Cotação Dolar " + dataHora.strftime('%d/%m/%y %H:%m') + ' '+ valorDolar)
```
